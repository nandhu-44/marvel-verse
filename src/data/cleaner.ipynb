```python
import pandas as pd
```

```python
movies_json_file = "./marvel_movies.json"
```

```python
df = pd.read_json(movies_json_file)
```

```python
df.columns
```

```python
# drop adult column as it is false for all rows
df.drop(columns=['adult'], inplace=True)
```

```python
# find all with no backdrop_path or no poster_path
df[(df['backdrop_path'].isnull()) | (df['poster_path'].isnull())]
```

```python
# drop these values
df.dropna(subset=['backdrop_path', 'poster_path'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# see where title and original_title are different
df[df['title'] != df['original_title']]
```

```python
# As no title is different from original_title, we can drop original_title
df.drop(columns=['original_title'], inplace=True)
```

```python
# Check where releae_date is an empty string
df[df['release_date'] == '']

# drop these values with empty release_date
df = df[df['release_date'] != '']
df.reset_index(drop=True, inplace=True)
df
```

```python
# save to cleaned_data.json file
df.to_json("cleaned_data.json", orient='records')
```
